```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
```

```python
df=pd.read_csv('marketing_campaign.csv',sep='\t')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Income'].fillna(df['Income'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe(include='all').T
```

```python
df['Dt_Customer']=pd.to_datetime(df['Dt_Customer'])
```

```python
df.dtypes
```

```python
df=pd.get_dummies(df,columns=['Education','Marital_Status'],drop_first=True)
```

```python
df
```

```python
df['Loyalty']=pd.to_datetime(datetime.date.today())-df['Dt_Customer']
```

```python
df.columns
```

```python
df['Loyalty']=df['Loyalty'].apply(lambda x: str(x).split()[0])
```

```python
df['Loyalty']=df['Loyalty'].astype(int)
#To bring in the loyalty based upon tenure 
```

```python
df_clust=df[['Year_Birth', 'Income', 'Kidhome', 'Teenhome',
       'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts',
       'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
       'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
       'NumStorePurchases','Education_Basic',
       'Education_Graduation', 'Education_Master', 'Education_PhD',
       'Marital_Status_Alone', 'Marital_Status_Divorced',
       'Marital_Status_Married', 'Marital_Status_Single',
       'Marital_Status_Together', 'Marital_Status_Widow',
       'Marital_Status_YOLO', 'Loyalty']]
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
```

```python
pca.fit(df_clust)
```

```python
var=pca.explained_variance_ratio_
```

```python
plt.plot(range(1,28),var.cumsum(),marker='o',linestyle='--')
plt.grid()
plt.show()
```

```python
pca=PCA(n_components=3)
```

```python
pca.fit(df_clust)
```

```python
df_pca=pca.fit_transform(df_clust)
```

```python
df_pca
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
wcss=[]
ss=[]

for i in range(2,30):
    k=KMeans(n_clusters=i)
    k.fit(df_pca)
    wcss.append(k.inertia_)
    ss.append(silhouette_score(df_pca,k.labels_))
```

```python
plt.plot(range(2,30),wcss,marker='o',linestyle='--')
plt.grid()
```

```python
plt.plot(range(2,30),ss,marker='o',linestyle='--')
```

```python
#We'll take number of clusters =2
```

```python
k=KMeans(n_clusters=2)
k.fit(df_pca)
```

```python
df['Clusters_Label']=k.labels_
```

```python
df.groupby(['Clusters_Label']).mean()
```

```python
df['Clusters_Name']=np.where(df['Clusters_Label']==1,'Budget Conscious','Affluent')
```

```python
df.groupby(['Clusters_Name']).agg({
    'Income': 'mean',
    'MntMeatProducts': 'mean',
    'MntFishProducts': 'mean',
    'MntSweetProducts': 'mean',
    'MntGoldProds': 'mean',
    'Loyalty':'mean'
})
```
